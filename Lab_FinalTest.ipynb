# Problem 1


```python
import pandas as pd

header = ['id','state','salary','experience']

data = pd.DataFrame([[1,'NY',57400,'Mid'],
 [2,'TX',None,'Entry'],
 [3,'NT',90000,'High'],
 [4,'VT',36900,'Entry'],
 [5,'TX', None,'Mid'],
 [6,'CA',76600,'High'],
 [7,'NY',85000,'High'],
 [8,'CA',None,'Entry'],
 [9,'CT',45000,'Entry']], columns = header)
print(data)

```

1. Handle missing values using the drop method

```python
#For cleaning using the Drop method
data.dropna()
```

2. Handle missing values using imputation

```python
#For cleaning using the imputation method
data.fillna(999.99)
```

3. (optional) Handle missing values using imputation considering the context of the record to reduce the bias in data.

```python
#Optional : Using the mean to added in the missing values of the dataframe
#find the mean of the column salary
salary_mean = data['salary'].mean()
print(salary_mean)

#Output to fill the missing values with the mean
data.fillna(salary_mean)
```

# **Problem 2**

```python
import pandas as pd

header = ['id','state','education']

data = pd.DataFrame([[1,'NY','High school'],
 [2,'TX',None],
 [3,'NT','High school'],
 [4,'VT','High school'],
 [5,'TX', 'None'],
 [6,'CA','College'],
 [7,'NY','High school'],
 [8,'CA',None],
 [9,'CT','College']], columns = header)
print(data)
```

How do you suggest that we handle the missing values?

```python
# Fill missing values with the mode (most frequent category)
data['education'].fillna(data['education'].mode()[0], inplace=True)
print(data)


```
